```python
import piplite
await piplite.install(["pandas", "matplotlib"])

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("dataset_part_2.csv")
data["Year"] = pd.to_datetime(data["Date"]).dt.year
annual_success = data.groupby("Year")["Class"].mean()

plt.figure(figsize=(10,6))
plt.plot(annual_success.index, annual_success.values, marker="o")
plt.title("Tendencia anual de éxito de lanzamiento")
plt.xlabel("Año")
plt.ylabel("Tasa de éxito promedio")
plt.grid(True)
plt.show()
```
